## 5.Modifying Data

### Renaming Columns 

```python
df.rename(columns={"sex_upon_intake": "sex"}, inplace=True)
```

### Changing Data Type

```python
df["age_upon_intake"]=df["age_upon_intake"].astype("string")
```

```python
df["datetime"]= pd.to_datetime(df["datetime"], errors='coerce')
```

### Adding a new column

```python
df["age_number"]=df["age_upon_intake"].str.split().str[0]
df["year/month"]=df["age_upon_intake"].str.split().str[1]
df["age_number"]=df["age_number"].astype("int")
```
